```python
import google.generativeai as genai
```

```python
genai.configure(api_key="Add Your API Key")

model = genai.GenerativeModel(
    model_name='gemini-2.5-flash'  #fastest + good quality
)
```

```python
SYSTEM_PROMPT = """
Your a helpful health assistant.

Rules:
-Always give a complete answer in 4-6 sentences,
-Never give one-line or incomplete answers.
-Explain causes clearly.
-Use simple languages.
-Avoid repeating words.

If unsure, say: consult a healthcare professioanl.

Example:
Q: What causes sor throat?
A: A sore throat is commonly caused by viral infections like the cold or flu. It can aalso happen due to allergies, dry air and irritation of smoke."""
```

```python
def safety_check(user_input):
  unsafe_keywords=[
      "suicide","kill myself","Overdose","prescribe","dosage for disease","emergency","heart attack","cancer cure"
  ]

  for words in unsafe_keywords:
    if words in user_input:
      return False
  return True
```

```python
def quick_reply(user_input):
  user_input_lower = user_input.lower()
  if "hello" in user_input_lower or "hi" in user_input_lower or "hey" in user_input_lower:
    return "Hi there! How can I assist you today?"
  if "how are you" in user_input_lower:
    return "I am just a program, but I'm ready to assist you"
  if "thank you" in user_input_lower or "thanks" in user_input_lower:
    return "Welcome"
  return None
```

```python
def chat_bot(user_input, history):
    # 1. Check for quick canned responses
    quick_res = quick_reply(user_input)
    if quick_res:
        return quick_res

    # 2. Safety check for sensitive topics
    if not safety_check(user_input):
        return "I cannot provide information on this topic. Please consult a healthcare professional or contact emergency services immediately if you are in danger."

    # 3. Generate response using Gemini
    try:
        prompt = f"{SYSTEM_PROMPT}\n\nUser Query: {user_input}\nAnswer:"
        response = model.generate_content(prompt)
        return response.text
    except Exception as e:
        return f"An error occurred: {str(e)}. Please try again."
```

```python
import gradio as gr

# Create the Chat Interface
demo = gr.ChatInterface(
    fn=chat_bot,
    title="AI Health Assistant",
    description="Ask general health questions. Note: Always consult a professional for medical advice.",
    examples=["What causes a sore throat?", "How can I improve my sleep?", "What are common symptoms of a cold?"]
)

if __name__ == "__main__":
    demo.launch(debug=True)
```
